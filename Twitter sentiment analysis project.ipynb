```python
training_data = pd.read_csv("C:\\Users\\HP\\Downloads\\0000000000002747_training_twitter_x_y_train.csv")
test_text = pd.read_csv("C:\\Users\\HP\\Downloads\\0000000000002747_test_twitter_x_test.csv")['text']
```

```python
texts = training_data['text']
Y_train = training_data['airline_sentiment']
```

```python
X =[]
X_test = []
for text in texts:
    X.append(word_tokenize(text))

for text in test_text:
    X_test.append(word_tokenize(text))
    
```

```python
stopwords = stopwords.words('english') + list(string.punctuation)
lemmatizer = WordNetLemmatizer()
def clean_doc(doc):
    clean_words=[]
    for word in doc:
        if word.lower() not in stopwords and ( word.isalpha()):
            clean_words.append( lemmatizer.lemmatize( word , simple(pos_tag([word])[0][1])).lower())
    return clean_words
```

```python
def simple(tag):
    if tag.startswith('N'):
        return wordnet.NOUN
    if tag.startswith('V'):
        return wordnet.VERB
    if tag.startswith('A'):
        return wordnet.ADV
    if tag.startswith('J'):
        return wordnet.ADJ
    else:
        return wordnet.NOUN
```

```python
X_train= [ ' '.join( clean_doc(doc)) for doc in X ]
X_test= [ ' '.join( clean_doc(doc)) for doc in X_test ]
X_test[0]
```

```python
Y_train = Y_train.replace({
    'negative' : 0,
    'neutral' : 1,
    'positive' : 2,
})
Y_train
```

```python
cv = CountVectorizer(max_features=3000)
cv.fit(X_train)
```

```python
cv.get_feature_names_out()
```

```python
X_train = cv.transform(X_train)
X_test = cv.transform(X_test)
```

```python
X_train = X_train.todense()
X_test = X_test.todense()
```

```python
import numpy as np
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import cross_val_score

X_train = np.asarray(X_train)
X_test = np.asarray(X_test)

model1 = MultinomialNB()
scores = cross_val_score(model1, X_train, Y_train)
print(scores.mean())
```

```python
from sklearn.linear_model import LogisticRegression
model2 = LogisticRegression(max_iter=2000,multi_class='ovr')
cross_val_score(model2,X_train,Y_train).mean()
```

```python
from sklearn.ensemble import RandomForestClassifier
model3 = RandomForestClassifier()
cross_val_score(model3,X_train,Y_train).mean()
```
